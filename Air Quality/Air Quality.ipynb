```python
#step-5
#Univariate Analysis(it shows one column at a time)
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df['PT08.S1(CO)'],kde=True)
plt.title("Distribution of 'PT08.S1(CO)'")
plt.show()
```

```python
#for outliers
#Boxplot
sns.boxplot(x=df[ 'PT08.S1(CO)'])
```

```python
#step-6
#bivariate Analysis(column vs column)
sns.scatterplot(x='PT08.S1(CO)', y='C6H6(GT)', data=df)
sns.lineplot(x='C6H6(GT)', y='RH', data=df)
```

```python
#step-7
#correction matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
```

```python
#step-8
#Time series analysis
df['Time'] = df['Time'].str.replace('.', ':', regex=False)

# Now combine and convert to datetime
df['datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], dayfirst=True)

# Set datetime as index
df.set_index('datetime', inplace=True)

# Plot
df['PT08.S1(CO)'].plot(figsize=(10, 5), title='PT08.S1(CO) Over Time')
```

```python
#step-9
#Time series Analysis
df['month']=df.index.month
df.groupby('month')['PT08.S1(CO)'].mean().plot(kind='bar')
```
